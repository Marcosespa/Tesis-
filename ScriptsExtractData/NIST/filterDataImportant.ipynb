```python
import os, re, subprocess
import pandas as pd

# 1) Carga tu dataframe (ya lo tienes como df). Si no:
# path = "/Users/marcosespana/Desktop/U/DatosTesis/ScriptsData/NIST/NIST-Cybersecurity-Publications.xlsx"
# df = pd.read_excel(path, engine="openpyxl")

# 2) Filtra a los que quieres (ya lo vienes haciendo)
# df = df[df['Stage']=="Final"]
# df = df[df['Series'].isin(["FIPS", "SP", "ITL Bulletin", "IR", "TN", "CSWP", "GCR", "AI"])]

BASE_OUT = "/Users/marcosespana/Desktop/U/DatosTesis/data/raw/NIST"

def norm_series(series: str) -> str:
    s = (series or "").strip()
    s = re.sub(r"\s+", "_", s)
    s = re.sub(r"[^A-Za-z0-9_.-]", "", s)
    return s or "Unknown"

def ensure_pdf_url(url: str) -> str:
    if not url:
        return ""
    u = url.strip()
    # Si ya termina en .pdf, usar tal cual
    if u.lower().endswith(".pdf"):
        return u
    # Heurística: muchos NIST PDFs están en nvlpubs y aceptan sufijo .pdf
    if "nvlpubs.nist.gov" in u and not u.endswith("/"):
        return u + ".pdf"
    return u  # fallback (curl descargará el recurso; si no es pdf, se verá en logs)

def curl_download(url: str, out_path: str) -> int:
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    cmd = [
        "curl", "-L", "--fail", "--retry", "3", "--retry-delay", "2",
        "-o", out_path, url
    ]
    return subprocess.call(cmd)

success, fail = 0, 0
for i, row in df.iterrows():
    series = norm_series(str(row.get("Series", "")))
    url = str(row.get("URL", "")).strip()
    if not url:
        url = str(row.get("CurrentURL", "")).strip()

    url_pdf = ensure_pdf_url(url)
    if not url_pdf:
        print(f"[skip] Sin URL en fila {i}")
        fail += 1
        continue

    # Nombre de archivo: usa PubID si existe; si no, basename del URL
    pub_id = str(row.get("PubID", "")).strip().replace(" ", "_")
    if pub_id:
        filename = pub_id + ".pdf" if not pub_id.lower().endswith(".pdf") else pub_id
    else:
        filename = os.path.basename(url_pdf.split("?")[0]) or f"file_{i}.pdf"
        if not filename.lower().endswith(".pdf"):
            filename += ".pdf"

    out_path = os.path.join(BASE_OUT, series, filename)
    if os.path.exists(out_path):
        # ya descargado
        continue

    rc = curl_download(url_pdf, out_path)
    if rc != 0:
        # Intento alterno: si probamos agregar .pdf y falló, intenta la URL sin .pdf (o viceversa)
        alt = url if url_pdf != url else (url + ".pdf")
        if alt != url_pdf:
            rc = curl_download(alt, out_path)

    if rc == 0 and os.path.getsize(out_path) > 0:
        success += 1
    else:
        fail += 1
        # Limpia archivo vacío
        if os.path.exists(out_path) and os.path.getsize(out_path) == 0:
            os.remove(out_path)
        print(f"[fail] {series} | {filename} | {url_pdf}")

print(f"Descargas OK: {success} | Fallidas: {fail}")
print(f"Destino base: {BASE_OUT}")
```
